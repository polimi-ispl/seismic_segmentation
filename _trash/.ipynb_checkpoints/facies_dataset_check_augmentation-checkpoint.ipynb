```python
import numpy as np
import _keras.data as D
import albumentations as albu
import utils as U
import matplotlib.pyplot as plt
```

```python
CLASS_NAMES = ["Basement", "SlopeMudA", "Deposit", "SlopeMudB", "SlopeValley", "Canyon"]
num_classes = len(CLASS_NAMES)
```

```python
dataset = "/nas/home/fpicetti/datasets/seismic_facies/train4.csv"
```

```python
d = np.load("/nas/home/fpicetti/datasets/seismic_facies/992_128_2_64/XZ000_p00.npy", allow_pickle=True).item()
```

```python
fig, axs = plt.subplots(1, 2, sharey=True, figsize=(16, 8))

axs[0].imshow(d['image'].squeeze(), cmap='gray')
axs[0].set_title('Image')
axs[0].axis('tight')

U.plot_label(np.argmax(d['mask'].squeeze(), axis=-1), fig, axs[1], num_classes, cmap='tab10')
axs[1].set_title('Mask')
axs[1].axis('tight')

plt.show()
```

```python
transforms = albu.Compose([
    albu.HorizontalFlip(),
    albu.Downscale(scale_min=0.5, scale_max=0.95, p=0.1),
])
```

```python
d_aug = albu.Compose([albu.HorizontalFlip(always_apply=True)])(**d)

fig, axs = plt.subplots(1, 2, sharey=True, figsize=(16, 8))

axs[0].imshow(d_aug['image'], cmap='gray')
axs[0].set_title('Image')
axs[0].axis('tight')

U.plot_label(np.argmax(d_aug['mask'], axis=-1), fig, axs[1], num_classes, cmap='tab10')
axs[1].set_title('Mask')
axs[1].axis('tight')

plt.show()
```

```python
d_aug = albu.Compose([albu.Downscale(scale_min=0.5, scale_max=0.95, always_apply=True)])(**d)

fig, axs = plt.subplots(1, 2, sharey=True, figsize=(16, 8))

axs[0].imshow(d_aug['image'], cmap='gray')
axs[0].set_title('Image')
axs[0].axis('tight')

U.plot_label(np.argmax(d_aug['mask'], axis=-1), fig, axs[1], num_classes, cmap='tab10')
axs[1].set_title('Mask')
axs[1].axis('tight')

plt.show()
```
