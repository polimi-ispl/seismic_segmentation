# synthetic wu

```python
img_dir = '/nas/home/fkong/data/faultdetect/synthetic_wu/'
result = {}
result['img_path'] = []
result['lbl_path'] = []
result['mode'] = []

nums = np.arange(200)
train_num = np.random.choice(200, size=180, replace=False)
image_path = os.path.join(img_dir, 'train', 'seis')
label_path = os.path.join(img_dir, 'train', 'fault')
for i in nums:    
    img_path = os.path.join(image_path, '%d.npy' % i)
    lbl_path = os.path.join(label_path, '%d.npy' % i)
    mode = 'train' if i in train_num else 'val'
    result['img_path'].append(img_path)
    result['lbl_path'].append(lbl_path)
    result['mode'].append(mode)

image_path = os.path.join(img_dir, 'validation', 'seis')
label_path = os.path.join(img_dir, 'validation', 'fault')
for i in range(20):
    img_path = os.path.join(image_path, '%d.npy' % i)
    lbl_path = os.path.join(label_path, '%d.npy' % i)
    mode = 'test'
    result['img_path'].append(img_path)
    result['lbl_path'].append(lbl_path)
    result['mode'].append(mode)
df = pd.DataFrame.from_dict(result)
df.to_csv('synthetic_wu/data.csv', index=False)
```

```python
ind = 10
seis = np.load('synthetic_wu/train/seis/%d.npy' % ind)[np.newaxis]
seis = (seis - seis.mean()) / seis.std()
fault = np.load('synthetic_wu/train/fault/%d.npy' % ind)[np.newaxis]
f"Number of inlines: {seis.shape[0]}, crosslines: {seis.shape[1]}, samples: {seis.shape[2]}"
```

```python
TIME=10
plt.figure(figsize=(12, 12))
plt.subplot(221)
plt.title('Original')
plt.imshow(seis[0, :,:, TIME].T, cmap='seismic')
plt.imshow(fault[0, :,:, TIME].T, cmap='Greys',  vmin=0, vmax=1, alpha=0.3)

seis = np.rot90(seis, k=1, axes=(0,1,2))
fault = np.rot90(fault, k=1, axes=(0,1,2))

plt.subplot(222)
plt.title('90')
plt.imshow(seis[0, :,:, TIME].T, cmap='seismic')
plt.imshow(fault[0, :,:, TIME].T, cmap='Greys',  vmin=0, vmax=1, alpha=0.3)

seis = np.rot90(seis, k=2, axes=(0,1,2))
fault = np.rot90(fault, k=2, axes=(0,1,2))

plt.subplot(223)
plt.title('180')
plt.imshow(seis[0,:,:, TIME].T, cmap='seismic')
plt.imshow(fault[0,:,:, TIME].T, cmap='Greys',  vmin=0, vmax=1, alpha=0.3)

seis = np.rot90(seis, k=3, axes=(0,1))
fault = np.rot90(fault, k=3, axes=(0,1))

plt.subplot(224)
plt.title('270')
plt.imshow(seis[:,:, TIME].T, cmap='seismic')
plt.imshow(fault[:,:, TIME].T, cmap='Greys',  vmin=0, vmax=1, alpha=0.3)
```
